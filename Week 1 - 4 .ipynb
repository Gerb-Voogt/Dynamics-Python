```python
import math, time, sys, os
import numpy as np
import matplotlib.pyplot as plt
```

```python
mass = 6 
t0 = 0
t1 = 20
dt = 0.1
v0 = 0
x0 = 0

t = np.linspace(t0, t1, 1 + round((t1-t0)/dt))
x_num = np.zeros(len(t))  
v_num = np.zeros(len(t))
x_num[0] = x0  #initial position 
v_num[0] = v0  #initial velocity

x_an = np.zeros(len(t))  
v_an = np.zeros(len(t))
x_an[0] = x0  #initial position 
v_an[0] = v0  #initial velocity
```

```python
def acceleration(t):
    if t > 10 and t < 20:
        return 0
    else:
        a = (3.5*t) / mass
        return a
```

```python
#Numerical Solution
for index in range(len(t)-1):
    a = acceleration(t[index])
    x_num[index+1] = x_num[index] + v_num[index]*dt
    v_num[index+1] = v_num[index] + a*dt


plt.plot(t, x_num, 'b', label='Pos')
plt.plot(t, v_num, 'r', label='Vel')
plt.legend()
```

```python
#Analytical Solution
for index in range(len(t)):
    if t[index] < 10 and t[index] < 20:
        v_an[index] = (0.5*3.5*t[index]**2)/mass
        x_an[index] = ((1/6)*3.5*t[index]**3)/mass
    else:
        v_an[index] =  v_an[99]
        x_an[index] =  v_an[index]*t[index] - 189

plt.plot(t, x_an, 'b', label='Pos')
plt.plot(t, v_an, 'r', label='Vel')
plt.legend()
```

```python
plt.xlabel('Time (s)')
plt.ylabel('x (m)')
plt.plot(t, x_num, 'b', label='Numerical')
plt.plot(t, x_an, 'r', label='Analytical')
plt.legend()
print(x_num[199])
print(v_num[199])
```
