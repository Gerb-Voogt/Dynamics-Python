```python
import math, time, sys, os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# All units in SI units
t0 = 0
t1 = 200
dt = 0.1
x0 = 0
v0 = 5
mass = 100
F = 12

t = np.linspace(t0, t1, 1 + round((t1 -t0)/dt))
x_num = np.zeros(len(t))
v_num = np.zeros(len(t))
x_num[0] = x0
v_num[0] = v0
```

```python
for i in range(len(t)-1):
    a = F/mass
    v_num[i+1] = v_num[i] + a*dt
    x_num[i+1] = x_num[i] + v_num[i]*dt

print('v = {}, x = {}'.format(v_num[-1], x_num[-1]))
```

```python
print('error = {}'.format(abs(3400 - x_num[-1])))
```

```python
plt.xlabel('x')
plt.ylabel('v')
plt.plot(x_num, v_num, label ='x vs v')
plt.legend()
```

```python
ans1 = np.interp(3000, x_num, v_num)
ans2 = np.interp(2000, x_num, t)

print('v @ 3000 m = {}, t @ 2000 m = {}'.format(ans1, ans2))
```

```python
print('error = {}'.format(abs(27.294688 - ans1)))
```
