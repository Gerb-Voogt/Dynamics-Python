```python
import math, time, sys, os
import matplotlib.pyplot as plt 
import numpy as np
```

```python
t0 = 0
t1 = 20
dt = 0.1

x0 = 0
v0 = 0

mass = 6

t = np.linspace(t0, t1, num=1+round((t1-t0)/dt))

x_num = np.zeros(len(t))
v_num = np.zeros(len(t))
x_num[0] = x0
v_num[0] = v0

x_an = np.zeros(len(t))
v_an = np.zeros(len(t))
v_an[0] = x0
x_an[0] = v0
```

```python
def acceleration(t):
    if t < 10:
        a = (3.5*t) / mass
    else:
        a = 0

    return a
```

```python
# Numerical
for index in range(len(t)-1):
    a = acceleration(t[index])
    x_num[index+1] = x_num[index] + v_num[index]*dt
    v_num[index+1] = v_num[index] + a*dt

plt.plot(t, x_num, 'b' ,label='x')
plt.plot(t, v_num, 'r' ,label='v')
plt.legend()

print(x_num[-1])
print(v_num[-1])
```
