```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
# Alles in SI eenheden (m, s, m/s)
t0 = 0
t1 = 10
dt = 0.1 
x0 = 0
v0 = 4

t = np.linspace(t0, t1, 1 + round((t1-t0)/dt))
x_an = np.zeros(len(t))  
v_an = np.zeros(len(t))

x_num = np.zeros(len(t))  
v_num = np.zeros(len(t))
x_num[0] = x0  #initial position 
v_num[0] = v0  #initial velocity
```

```python

for index in range(len(t)):
    a = 0
    v_an[index] = v0 + a*t[index]
    x_an[index] = x0 + v0*t[index] + 0.5*a*t[index]**2

print(x_an[-1])

for index in range(len(t)-1):
    a = 0
    x_num[index+1] = x_num[index] + v_num[index]*dt
    v_num[index+1] = v_num[index] + a*dt

# print('x_num = {}'.format(x_num))
# print('v_num = {}'.format(v_num))
print(x_num[80])
```

```python
plt.figure(1)
plt.plot(t, x_num)
plt.plot(t, x_an, 'y--')

plt.figure(2)
plt.plot(t, x_num-x_an)
```
