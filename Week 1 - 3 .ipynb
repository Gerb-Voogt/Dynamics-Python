```python
import math, time, sys, os
import matplotlib.pyplot as plt
import numpy as np
```

```python
mass = 6 
t0 = 0
t1 = 10
dt = 0.1
v0 = 0
x0 = 0

t = np.linspace(t0, t1, 1 + round((t1-t0)/dt))
x_num = np.zeros(len(t))  
v_num = np.zeros(len(t))
x_num[0] = x0  #initial position 
v_num[0] = v0  #initial velocity

x_an = np.zeros(len(t))  
v_an = np.zeros(len(t))
x_an[0] = x0  #initial position 
v_an[0] = v0  #initial velocity
```

```python
def acceleration(t):
    a = np.sin(t) / mass
    return a
```

```python
for index in range(len(t)-1):
    a = acceleration(t[index])
    x_num[index+1] = x_num[index] + v_num[index]*dt
    v_num[index+1] = v_num[index] + a*dt

plt.plot(t, x_num, 'b', label='Pos')
plt.plot(t, v_num, 'r', label='Vel')
plt.legend()

print('Pos = {}'.format(x_num[80]))
print('Vel = {}'.format(v_num[80]))
```

```python
# Analytical Solution
for index in range(len(t)):
    v_an[index] = (1/mass) - (np.cos(t[index]))/mass
    x_an[index] = t[index]/mass - (np.sin(t[index]))/mass

plt.plot(t, x_an, 'b', label='Pos')
plt.plot(t, v_an, 'r', label='Vel')
plt.legend()

print('Pos = {}'.format(x_an[-1]))
print('Vel = {}'.format(v_an[-1]))
print('diff = {}'.format(abs(x_num[-1] - x_an[-1])))
```
